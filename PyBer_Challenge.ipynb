```python
# Import dependancies
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
# Files to load
city_data_to_load = os.path.join('.', 'Resources', 'city_data.csv')
ride_data_to_load = os.path.join('.', 'Resources', 'ride_data.csv')
```

```python
# Read the city data file and store it in a pandas DataFrame.
city_data_df = pd.read_csv(city_data_to_load)
city_data_df.head()
```

```python
# Read the ride data file and store it in a pandas DataFrame.
ride_data_df = pd.read_csv(ride_data_to_load)
ride_data_df.head()
```

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the DataFrame
pyber_data_df.head()
```

# Challenge 1: Get a Summary DataFrame

```python
# 1. get total rides for each city type
total_rides_by_type = pyber_data_df.groupby(['type']).count()['ride_id']
total_rides_by_type
```

```python
# 2. get total drivers for each city type
total_drivers_by_type = city_data_df.groupby(['type']).sum()['driver_count']
total_drivers_by_type
```

```python
# 3.get the total amount of fare for each city type
total_fare_by_type = pyber_data_df.groupby(['type']).sum()['fare']
total_fare_by_type
```

```python
# 4. get the average fare per ride for each city type
avg_fare_per_ride = total_fare_by_type /total_rides_by_type
avg_fare_per_ride
```

```python
# 5. get the average fare per ride for each city type
avg_fare_per_driver = total_fare_by_type /total_drivers_by_type
avg_fare_per_driver
```

```python
# Summary Table

summary = {
    'Total Rides': total_rides_by_type,
    'Total Drivers': total_drivers_by_type,
    'Total Fares': total_fare_by_type,
    'Average Fare per Ride':avg_fare_per_ride,
    'Average Fare per Driver': avg_fare_per_driver
}

pyder_ride_summary_df = pd.DataFrame(summary)
pyder_ride_summary_df
```

```python
# 7. delete the index name
pyder_ride_summary_df.index.name = ''
pyder_ride_summary_df
```

```python
# 8. Format the columns
pyder_ride_summary_df['Total Rides'] = pyder_ride_summary_df['Total Rides'].map('{:,}'.format)
pyder_ride_summary_df['Total Drivers'] = pyder_ride_summary_df['Total Drivers'].map('{:,}'.format)
pyder_ride_summary_df['Total Fares'] = pyder_ride_summary_df['Total Fares'].map('${:,.2f}'.format)
pyder_ride_summary_df['Average Fare per Ride'] = pyder_ride_summary_df['Average Fare per Ride'].map('${:,.2f}'.format)
pyder_ride_summary_df['Average Fare per Driver'] = pyder_ride_summary_df['Average Fare per Driver'].map('${:,.2f}'.format)

pyder_ride_summary_df
```

# Challenge 2 Create a line plt that shows the sum of the fares for each type of city

```python
# 1. Rename columns
# city	date	fare	ride_id	driver_count	type
pyber_data_df = pyber_data_df.rename(columns={
    'city': 'City',
    'date': 'Date',
    'fare': 'Fare',
    'ride_id': 'Ride ID',
    'driver_count': 'No. Drivers',
    'type': 'City Type'
})

pyber_data_df
```

```python
# 2. set the index to the 'Date' column
pyber_data_df.set_index(pyber_data_df['Date'], inplace=True)
pyber_data_df
```

```python
# 3. make new dateframe for fares and include only Date, City Type, and Fare columns
column_names = ['Date', 'City Type', 'Fare']
pyber_cities_fare = pyber_data_df[column_names].copy()
pyber_cities_fare
```

```python
# 4. drop the extra columns
pyber_cities_fare.drop(['Date'], axis=1, inplace=True)
pyber_cities_fare
```

```python
pyber_cities_fare.index = pd.to_datetime(pyber_data_df.index)
pyber_cities_fare.head()
```

```python
pyber_cities_fare.info()
```

```python
# 7. calculate the sum of the fares by the type of the city and date using groupby to create a series
sum_fare_by_type = pyber_cities_fare.groupby(['City Type', 'Date']).sum()['Fare']
sum_fare_by_type
```

```python
# 7b. Alternately you can create a DataFrame by using double brackets
sum_fare_by_type = pyber_cities_fare.groupby(['City Type', 'Date']).sum()[['Fare']]
sum_fare_by_type
```

```python
# convert using the pd.DataFrame method
sum_fare_by_type = pd.DataFrame(sum_fare_by_type)
sum_fare_by_type
```

```python
# 9. reset index so we can pivot method in pandas
sum_fare_by_type = sum_fare_by_type.reset_index()
sum_fare_by_type
```

```python
# 10. create a pivot table with the data as the index and columns as the type with the fare in each row
sum_fare_by_type_pivot = sum_fare_by_type.pivot(index='Date', columns='City Type')['Fare']
sum_fare_by_type_pivot.head()
```

```python
sum_fare_by_type_pivot.info()
```

# Create a line chart that shows from Jan 1, 2018 to April 28, 2018

```python
fare_Jan_April = sum_fare_by_type_pivot.loc['2019-01-01': '2019-04-28']
fare_Jan_April.head()
```

```python
# 12. create a new dataframe and use resample by week and get sum of the fares for each week
weekly_fares_df = fare_Jan_April.resample('W').sum()
weekly_fares_df
```

```python
# 13. use axes plotting to add labels and titles
from matplotlib import style
style.use('fivethirtyeight')
ax = weekly_fares_df.plot(figsize=(20,6))
# add a title
ax.set_title('Ttoal Fare by City type')
# add a x- and y-axis label
ax.set_xlabel('Month')
ax.set_ylabel('Fare ($USD)')
plt.savefig('analysis/Challenge_fare_summary.png')
plt.show()
```
